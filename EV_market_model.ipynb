```python
sse=[]
for k in range (1,10):
    kmean=KMeans(n_clusters=k)
    kmean.fit(df)
    sse.append(kmean.inertia_)
```

```python
sse
```

```python
plt.title("cluster of EV Data{Elbow point method}")
plt.plot([i for i in range(1,10)],sse)
plt.show()
```

```python
cluster = KMeans(n_clusters=3)
df['Cluster']=cluster.fit_predict(df)
```

```python
df.head()
```

```python
def plotseg(clus,Y, labels):  
  plt.scatter(clus[Y==0,0], clus[Y==0,1], s=10, c='green', label='Cluster 1')
  plt.scatter(clus[Y==1,0], clus[Y==1,1], s=10, c='red', label='Cluster 2')
  plt.scatter(clus[Y==2,0], clus[Y==2,1], s=10, c='blue', label='Cluster 3')

  plt.title('Customer_segments')
  plt.xlabel(labels[0])
  plt.ylabel(labels[1])
  plt.show()
```

```python
clus = df.loc[:,["AccelSec","Price"]].values
kmeans = KMeans(n_clusters=3, init='k-means++')
Y = kmeans.fit_predict(clus)
plotseg(clus, Y, ["AccelSec","Price"]) 
```

```python
clus = df.loc[:,["TopSpeed_KmH","Price"]].values
kmeans = KMeans(n_clusters=3, init='k-means++')
Y = kmeans.fit_predict(clus)
plotseg(clus, Y, ["TopSpeed_KmH","Price"]) 
```
